```python
from src.main.DatasetManager import DatasetManager
from src.main.models.ClusteredUserBasedCF import ClusteredUserBasedCF
```

```python
dirname = '/home/damian/Pulpit/Warsztat/CoML/recomm_project/data/'

ml100k_filename = dirname + 'ml-100k/u.data'
ml100k_item_filename = dirname + 'ml-100k/u.item'

# load data
dm = DatasetManager()
data = dm.load_csv(ml100k_filename, realative=False)
train, test = dm.train_test_split(data, shuffle=False)
items = dm.load_csv(ml100k_item_filename, delimeter='|', encoding='ISO-8859-1', realative=False)

# split to train and test set 
user_item = dm.transform_to_user_item_mat(train, verbose=True)
user_item_test = dm.transform_to_user_item_mat(test, verbose=True)
```

```python
# train 
model = ClusteredUserBasedCF(verbose=True)
model.fit(user_item)
```

```python
# test
model.test(user_item_test)
```

```python
# get recommendations 
recomm = model.recommend_all()
```

```python
def print_user_recommendations(items, recommendations, watched_movies, users_num=10):
    for i in range(users_num):
        print("====== User with id: {}".format(i))
        print("\tWatched:")
        for idx, watched_obj in enumerate(watched_movies[i]):
            id, rating = watched_obj['item_id'], watched_obj["rating"]
            title = items[id][1]
            print("\t{}. id: {}, rating: {}, title: {}".format(idx + 1, id, rating, title))

        print("\tRecommended:")
        for idx, recomm_obj in enumerate(recommendations[i]):
            id, rating = recomm_obj['item_id'], recomm_obj["pred_rating"]
            title = items[id][1]
            print("\t{}. id: {}, pred_rating: {}, title: {}".format(idx + 1, id, rating, title))

def get_watched_movies(user_item_mat):
    watched = []
    
    for i in range(len(user_item_mat)):
        watched.append([])
        for j in range(len(user_item_mat[0])):
            if user_item_mat[i][j] > 0:
                watched[i].append({"item_id": j, "rating": user_item_mat[i][j]})
            
    for i in range(len(watched)):
        watched[i].sort(key=lambda o: o["rating"], reverse=True)
    
    return watched
    
```

```python
print("User based: ")
watched_movies = get_watched_movies(user_item)
print_user_recommendations(items, recomm, watched_movies, users_num=3)
```
